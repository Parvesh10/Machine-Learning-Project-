Data Set Information:

The dataset includes 244 instances that regroup & data of two regions of Algens namely the Bejaia repon located in the northeast of Algeria and the Sid be-actes regon located in the northwest of Algeria.

122 instances for each region

The period from June 2012 to September 2012. The dataset indudes 11 atribues and 1 output attribue(class) The 244 instances have been classified into fire(138 classes) not fire (106 classes) classes.

Attribute Informations:

1. Date: (DD/MM/YYYY) Day month (june' to 'september), year (2012) Weather data observations
2. Temp: temperature noon demperature mas) in Celsius degrees: 2242
3. RH: Relative Humidity in 21 to 90
4. Ws Wind speed in kmitt 6 to 25
5. Rain: total day in mm 0 to 16.8 FWI Components
6. Fine Fuel Moisture Code (FFMC) inder from the PWM system 256 to 92.5
7. Duff Moisture Code (DMC) indes from the FW system: 1.3 to 659
8. Drought Code (DC) index from the PWI system: 7 to 2204
9. Initial Spread Index (5) index from the PWI system: 0 to 18.5
10. Buildup Index (BU) index from the PWI system: 1.1 to 68
11. Fire Weather Inder (FW) Index: 0 to 31.1
12. Classes too classes, namely Fire and not Fee
